__Q1) Explain GET and POST methods?__

__Ans)__ GET and POST are HTTP methods used to request and send data between a client and a server. Here's an explanation of each:

* __GET:__ This method is used to request a resource from a server. When a client sends a GET request, it asks the server to return a representation of the specified resource. This representation can be in various formats such as HTML, JSON, or XML. GET requests are used for operations that don't change the state of the server or the data it holds. For example, when a user enters a URL in a web browser, the browser sends a GET request to the server to retrieve the web page.

* __POST:__ This method is used to submit data to a server to create or update a resource. When a client sends a POST request, it includes a payload that contains the data to be sent to the server. This payload can be in various formats such as JSON, XML, or form data. POST requests are used for operations that change the state of the server or the data it holds. For example, when a user submits a form on a web page, the browser sends a POST request to the server with the form data.

__Q2) Why is request used in Flask ?__

__Ans)__ In Flask, the request object is used to represent an HTTP request sent by a client to the server. The request object contains information about the request such as the method used (GET, POST, etc.), the URL requested, and any data sent in the request body.

The request object is useful in Flask because it allows you to access the data sent by the client in a request. For example, if a user submits a form on a web page, the data entered in the form can be accessed using the request object in Flask.

__Q3) Why is redirect() used in Flask ?__

__Ans)__ The redirect() function is used in Flask to redirect the user's browser to a different URL.

In web applications, it is common to redirect users to different pages or routes based on their actions or authentication status. For example, after a user logs in to a website, they may be redirected to their account dashboard. Alternatively, if a user tries to access a protected resource without being authenticated, they may be redirected to a login page.

In Flask, the redirect() function makes it easy to implement such redirects. When the function is called, it creates a response object that instructs the user's browser to make a new request to the specified URL.

__Q4)What are templates in Flask? Why is the render_template() function used?__

__Ans)__  Templates in Flask are files that contain __HTML, CSS, and other static content__, as well as dynamic content generated by a Flask application. Templates are used to create the user interface of web applications, which can vary based on user input or other factors.

In Flask, templates are typically stored in a templates folder in the application directory. Flask uses the Jinja2 templating engine to render templates and generate dynamic content.

The render_template() function in Flask is used to render a template and generate the corresponding HTML that is returned to the client's browser. The function takes the name of the template file and any additional data to be passed to the template as arguments.

```python
from flask import Flask,request ,render_template , jsonify

app = Flask(__name__)

@app.route('/postman_action',methods=['POST'])
def math_ops1():
    if(request.method == 'POST'):
        ops = request.json['operation']
        num1 = int(request.json['num1'])
        num2 = int(request.json['num2'])
        if ops == 'add':
            r = num1+num2
            result = "The sum of " + str(num1) + 'and ' + str(num2) + "is " + str(r)
        if ops == 'subtract':
            r = num1-num2
            result = "The subtract of " + str(num1) + 'and ' + str(num2) + "is " + str(r)
        if ops == 'multiply':
            r = num1*num2
            result = "The multiply of " + str(num1) + 'and ' + str(num2) + "is " + str(r)
        if ops == 'divide':
            r = num1/num2
            result = "The divide of " + str(num1) + 'and ' + str(num2) + "is " + str(r)
            
        return jsonify(result)

if __name__=="__main__":
    app.run(host="0.0.0.0")
```
